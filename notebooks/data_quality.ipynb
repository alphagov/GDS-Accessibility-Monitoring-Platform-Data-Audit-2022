```python
import warnings

warnings.simplefilter(action="ignore", category=FutureWarning)


import pandas as pd
import sqlite3
import psycopg2
import sys
from sqlalchemy import create_engine
```

```python
url = "postgresql+psycopg2://admin:secret@localhost:5432/accessibility_monitoring_app"
engine = create_engine(url)
```

```python
pd.set_option("display.max_columns", None)


df = pd.read_sql("SELECT * FROM public.cases_case;", engine)
filtered_df = df[df["created_by_id"].notnull()]
filtered_df
```

```python
pd.set_option("display.max_columns", None)


list_of_series = []

for col in filtered_df:
    isnull = filtered_df[filtered_df[col].isnull()].shape[0] / df.shape[0]
    notnull = filtered_df[filtered_df[col].notnull()].shape[0] / df.shape[0]
    describe = filtered_df[col].describe()
    to_append = pd.Series([isnull, notnull], index=["isnull_perc", "notnull_perc"])
    describe = describe.append(to_append)
    list_of_series.append(describe.to_frame(col))

df_quality = pd.concat(list_of_series, axis=1)
df_quality
```
